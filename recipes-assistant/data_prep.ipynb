```python
import pandas as pd
```

```python
df = pd.read_csv('recipes.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.dropna(subset=["total_time"])
```

```python
import re

def to_minutes(s):
    hours = re.search(r'(\d+)\s*hrs?', s)
    mins = re.search(r'(\d+)\s*mins?', s)

    h = int(hours.group(1)) if hours else 0
    m = int(mins.group(1)) if mins else 0

    return h * 60 + m

df["total_minutes"] = df["total_time"].apply(to_minutes)
```

```python
df_filtered = df[df["total_minutes"] <= 60]
```

```python
df_filtered.shape
```

```python
df_filtered.head(2)
```

```python
df_filtered = df_filtered[["recipe_name", "total_time", "servings", "ingredients", "directions", "nutrition"]]
```
